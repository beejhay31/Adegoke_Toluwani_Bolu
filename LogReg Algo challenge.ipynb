# Function definitions

```python
#initializing parameters
def initialize_theta(dimension):
    theta_others=np.full((dimension, 1), 0.01)
    theta_nut=0.0
    return theta_others,theta_nut
```

```python
#a sigmoid function is otherwise known as a logistic function and the formula is g(z) = 1/1+e**-z
def sigmoid(Z):
    g=np.zeros(np.size(Z))
    g = 1/(1+np.exp(-Z))
    return g
```

```python
#simplified cost function and gradient descent(to minimize cost function to achieve fitting parameters)
def cost_and_grad(theta_others, theta_nut, X_train, Y_train):
    Z=np.dot(theta_others.T, X_train)+theta_nut
    g=sigmoid(Z)
    loss=-Y_train*np.log(g)-(1-Y_train)*np.log(1-g)
    m=np.size(X_train, 1)
    cost = (np.sum(loss))/m
    
    #applying gradient descent to minimize cost function in order to fit parameter(theta)
    d_theta_others=(np.dot(X_train,((g-Y_train).T)))/m
    d_theta_nut=(np.sum(loss))/m
    grads= {'d_theta_others':d_theta_others, 'd_theta_nut':d_theta_nut}
    return cost, grads
```

```python
#fuction to update cost, parameters and plot the logistic regression graph
def update(theta_others, theta_nut, X_train, Y_train, alpha, num_iter):
    cost_l = []
    cost_l2 = []
    index = []
    
    for i in range(num_iter):
        cost, grads = cost_and_grad(theta_others, theta_nut, X_train, Y_train)
        cost_l.append(cost)
        
        
        #formula:repeat{theta=theta-learning_rate*(d/d_theta)cost}
        #where: (d/d_theta)cost = 1/m*summation(sigmoid-y)x for all i=1 to m
        theta_others=theta_others-alpha*grads['d_theta_others']
        theta_nut=theta_nut-alpha*grads['d_theta_nut']
        
        #initiating the steps of the cost i.e for 10, it will be in '0, 10, 20'
        if i % 10 == 0:
            cost_l2.append(cost)
            index.append(i)
            print ("cost after % i: % f" %(i, cost))
            
   #plotting the cost against iterations graph
    parameters = {"theta_others":theta_others, "theta_nut":theta_nut}
    plt.plot(index, cost_l2)
    plt.xticks(index, rotation='vertical')
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.title('Logistic Regression')
    plt.show()
    return parameters, grads, cost_l
```

```python
#building the prediction function
#z=transpose(theta).X
#prediction = 1 if sigmoid(Z)>=0.5
#else prediction = o

def predict(theta_others, theta_nut,  X_test):
    Z=sigmoid(np.dot(theta_others.T, X_test)+theta_nut)
    Y_pred=np.zeros((1, np.size(X_test, 1)))
    n=np.size(Z, 1)
    
    for i in range (n):
        if Z[0, i]>=0.5:
            Y_pred[0, i]=1
        else:
            Y_pred[0,i]=0
    return Y_pred
```

```python
#Building the logistic regression function by applying previously built functions.

def logreg(X_train, Y_train, X_test, Y_test, alpha, Iter):
    
    dimension=np.size(X_train, 0)
    
    theta_others, theta_nut=initialize_theta(dimension)
    
    parameters, grads, cost_l = update(theta_others, theta_nut, X_train, Y_train, alpha,Iter)
    
    Y_pred_test=predict(parameters['theta_others'], parameters['theta_nut'], X_test)
    Y_pred_train=predict(parameters['theta_others'], parameters['theta_nut'], X_train)
    
    
    
    print('train accuracy: {} %'.format(100-np.mean(np.abs(Y_pred_train-Y_train))*100))
    print('test accuracy: {} %'.format(100-np.mean(np.abs(Y_pred_test-Y_test))*100))

logreg(X_train, Y_train, X_test, Y_test, alpha=0.25, Iter=150)
```
